```python
import numpy as np
import pandas as pd
# matplolib
import matplotlib.pyplot as plt
# seaborn
import seaborn as sns
```

```python
# Data analysis of 'chocolate protfolio project' dataset

df1 = pd.read_excel('/content/choclate protfolio project - 11.xlsx',header=0)
df1.head(10)
```

```python
# Pivot Table

pd.pivot_table(index='Sales Person' , columns= ['Geography','Product'], values='Amount' ,aggfunc='sum',data=df1)
```

```python
pd.pivot_table(index=['Sales Person','Product'], columns= 'Geography', values='Amount' ,aggfunc='sum',data=df1)
```

```python
# seaborn library dataset 'flights' loading and visualisation on this dataset

df=sns.load_dataset('flights')
df.tail()
```

```python
df[df['passengers']==df['passengers'].max()][['year','month']]
```

```python
flight=df.pivot_table(index='year',values='passengers',columns='month')
flight
```

```python
sns.heatmap(flight)
```
