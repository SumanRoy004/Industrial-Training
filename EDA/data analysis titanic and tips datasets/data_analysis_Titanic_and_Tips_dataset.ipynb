```python
# matplolib
import matplotlib.pyplot as plt
# seaborn
import seaborn as sns
```

```python
# Titanic data visualisation in the form of chart

df = pd.read_csv('/content/train (4).csv')
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()/len(df)*100
```

```python
# handling missing values
#1. >50% --- col drop
#2. <50% --- value replace (mean/median/random)
#3. 2-5% --- row delete
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
# mean --
# median --- outliers
```

```python
df['Age'].median()
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()/len(df)*100
```

```python
# GroupBy

df.groupby('Survived')['Age'].mean()
```

```python
df.groupby('Pclass')['Fare'].mean()
```

```python
df.groupby('Survived')['Sex'].value_counts()
```

```python
# avg fare --- Embarked, Pclass

df.groupby(['Embarked','Pclass'])['Fare'].mean()
```

```python
# pie chart using pandas
df['Embarked'].value_counts().plot(kind='pie')
```

```python
# pie chart using matplotlib
_=plt.pie(df['Embarked'].value_counts(),labels=['S','C','Q'],autopct='%0.1f%%',
               colors=['#EB0B3E','#96EB0B','#0B82EB'],shadow=True,explode=[0,0.09,0])
```

```python
# seaborn library dataset 'tips' loading and visualisation on this dataset

df4 = sns.load_dataset('tips')
df4.head()
```

```python
# scatterplot using matplotlib
plt.scatter(x='total_bill',y='tip',data=df4,color='r',edgecolors='black')
```

```python
# scatterplot using seaborn
sns.scatterplot(x='total_bill',y='tip',data=df4,hue='sex',style='smoker',size='time')
```

```python
df4['time'].unique()
```

```python
# boxplot using seaborn
sns.boxplot(x='day',y='total_bill',data=df4)
```

```python
# barplot using seaborn
sns.barplot(x='day',y='total_bill',data=df4,estimator='sum')
```

```python
# histogram using seaborn
sns.displot(x='total_bill',data=df4)
```

```python
# density plot using seaborn
sns.displot(x='total_bill',data=df4,kind='kde',hue='sex')  # kde=Kernel Density Estimation
```

```python
# swarmplot using seaborn
sns.catplot(x='time',y='total_bill',hue='sex',kind='swarm',col='day',data=df4,aspect=.5)
```
