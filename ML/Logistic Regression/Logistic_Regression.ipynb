```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/iphone_purchase_records.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Purchase Iphone'].value_counts(normalize=True)
```

```python
# Replacing the 'Male' and 'Female' values by 0 and 1 respectively
df['Gender']=df['Gender'].replace({'Female':1,'Male':0})
```

```python
df.head()
```

```python
x=df.iloc[:,0:3].values
y=df.iloc[:,-1].values
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
# Model Creation
from sklearn.linear_model import LogisticRegression
logit=LogisticRegression()
logit.fit(x_train,y_train)
```

```python
y_pred=logit.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Performance Measurement using accuracy and Performance Metrics
from sklearn.metrics import accuracy_score,confusion_matrix
ACC=accuracy_score(y_test,y_pred)
print(ACC)
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
