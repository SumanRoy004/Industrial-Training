```python
df.dropna(inplace=True)
```

```python
df_final = pd.get_dummies(df,drop_first=True)
```

```python
df_final.shape
```

```python
x = df_final.drop('median_house_value', axis=1)
y = df_final['median_house_value']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,  random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Creating the ANN
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann_clf = Sequential()
# input layer
ann_clf.add(Dense(activation = 'relu', input_dim=12, units=150, kernel_initializer='uniform'))
```

```python
# first hidden layer
ann_clf.add(Dense(activation = 'relu', units=150, kernel_initializer='uniform'))
# second hidden layer
ann_clf.add(Dense(activation = 'relu', units=150, kernel_initializer='uniform'))
# third hidden layer
ann_clf.add(Dense(activation = 'relu', units=150, kernel_initializer='uniform'))
# fourth hidden layer
ann_clf.add(Dense(activation = 'relu', units=150, kernel_initializer='uniform'))
# fifth hidden layer
ann_clf.add(Dense(activation = 'relu', units=150, kernel_initializer='uniform'))
# sixth hidden layer
ann_clf.add(Dense(activation = 'relu', units=150, kernel_initializer='uniform'))
# seventh hidden layer
ann_clf.add(Dense(activation = 'relu', units=150, kernel_initializer='uniform'))
# eighth hidden layer
ann_clf.add(Dense(activation = 'relu', units=150, kernel_initializer='uniform'))
# ninth hidden layer
ann_clf.add(Dense(activation = 'relu', units=150, kernel_initializer='uniform'))
# tenth hidden layer
ann_clf.add(Dense(activation = 'relu', units=150, kernel_initializer='uniform'))
# output Layers
ann_clf.add(Dense(activation ='linear', units=1, kernel_initializer='uniform'))

# model compile
ann_clf.compile(optimizer='adam',loss='mse', metrics=['mae'])
```

```python
# model Train
ann_clf.fit(x_train,y_train,validation_split=0.33,batch_size=32,epochs=160)
```

```python
y_pred=ann_clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
